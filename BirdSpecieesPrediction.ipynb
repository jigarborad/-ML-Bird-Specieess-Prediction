```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
import keras
from keras.preprocessing.image import ImageDataGenerator
import cv2
from PIL import Image
import os
from keras.preprocessing import image
from sklearn.preprocessing import LabelBinarizer
from tensorflow.keras.optimizers import Adam
```

```python
!pip install adam
```

```python
dir = 'Bird Speciees Dataset'
root_dir = os.listdir(dir)
image_list , label_list = [],[]
```

```python
from posix import listdir
for directory in root_dir:
  for files in listdir(f"{dir}/{directory}"):
    image_path = f"{dir}/{directory}/{files}"
    image_file = cv2.imread(image_path)
    image_file = image.img_to_array(image_file)
    image_list.append(image_file)
    label_list.append(directory)
```

```python
count = pd.DataFrame(label_list).value_counts()
count
```

```python
image_list[0].shape
```

```python
label_list = np.array(label_list)
label_list.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(image_list,label_list, test_size=0.2, random_state = 0) 
```

```python
X_train = np.array(X_train,dtype=np.float16)/255.0
X_test = np.array(X_test,dtype=np.float16)/255.0
X_train = X_train.reshape(-1,224,224,3)
X_test = X_test.reshape(-1,224,224,3)
```

```python
lb = LabelBinarizer()
y_train = lb.fit_transform(y_train)
y_test = lb.fit_transform(y_test)
```

```python
X_train,X_val,y_train,y_val = train_test_split(X_train,y_train, test_size=0.2, random_state = 0) 
```

```python
from keras.layers import Dropout
```

```python
cnn = tf.keras.models.Sequential()
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32,padding="same" , kernel_size=(3,3), activation='relu', input_shape=(224, 224, 3)))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=(2,2), strides=2))
cnn.add(Dropout(0.25))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=48 ,padding="same", kernel_size=(3,3), activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=(2,2)))
cnn.add(Dropout(0.25))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=16 ,padding="same", kernel_size=(3,3), activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=(2,2)))
cnn.add(Dropout(0.25))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(units=32, activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units=len(root_dir), activation='softmax'))
```

```python
cnn.compile(loss= 'categorical_crossentropy', optimizer =Adam(0.0005), metrics = ['accuracy'])
```

```python
history = cnn.fit(x = X_train, y = y_train, batch_size = 128, epochs = 50, validation_data=(X_val, y_val))
```

```python
scores = cnn.evaluate(X_test, y_test)
print(f"Test Accuracy: {scores[1]*100} %")
```
